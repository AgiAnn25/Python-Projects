## **2. Dataset Column Overview**

1. PassengerId :- Unique ID for each passenger

2. Survived :- Survival (0 = No, 1 = Yes)

3. Pclass :- Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd)

4. Name :- Passenger name

5. Sex :- Gender

6. Age :- Age in years

7. SibSp : Siblings/Spouses aboard

8. Parch :- Parents/Children aboard

9. Ticket :- Ticket Number

10. Fare :- Fare Paid

11. Cabin :- Cabin Number

12. Embarked :- Port of Embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

## **5. Handling with Missing Values**

```python
Titanic_data = Titanic_data.drop(columns='Cabin', axis = 1)
```

```python
Titanic_data['Age'] = Titanic_data['Age'].fillna(Titanic_data['Age'].mean())
```

```python
Titanic_data.info()
```

```python
Titanic_data['Embarked']=Titanic_data['Embarked'].fillna(Titanic_data['Embarked'].mode()[0])
```

```python
Titanic_data.info()
```

```python
Titanic_data.isnull().sum()
```

In the Titanic dataset, some columns had missing data. The Cabin column had too many missing values, so it was removed. For the Age column, missing values were filled with the average age. This helps to keep the age data realistic. The Embarked column had a few missing values, so they were filled with the most common value in that column.

## **7. Feature Engineering**

```python
Titanic_data.head(5)
```

## **8. Encoding the Categorical Columns**

```python
Titanic_data['Sex'].value_counts()
```

```python
Titanic_data['Embarked'].value_counts()
```

```python
Titanic_data.replace({'Sex':{'male': 0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}})
```

The Sex column is converted into numbers: male = 0, female = 1. The Embarked column is also changed into numeric values: C = 0, Q = 1, and S = 2. This makes the dataset suitable for machine learning models, which require numerical input instead of text.



```python
x = Titanic_data.drop(columns=['Name','Ticket'],axis = 1)
```

```python
x.head(5)
```

```python
x = x.replace({'Sex':{'male': 0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}})
```

```python
x.head(5)
```

Removing Name and Ticket columns helps clean the data. These columns contain text that doesn’t add much value for prediction. Keeping only useful columns makes the dataset better for building machine learning models.

## **9. Separating the Features and Targets**

```python
X = x.drop(columns=['Survived'])
```

```python
x.head(5)
```

```python
Y = x['Survived']
```

```python
Y
```

To prepare the data for model training, the features and target were separated. The Survived column, which is the target variable, was stored in y. All other columns that help predict survival were stored in X. This separation allows the model to learn patterns from the input features and predict the survival outcome.



## **10. Splitting the data into training data & Test data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state=2)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

The dataset was split into training and testing sets using the train_test_split function. 70% of the data was used for training the model (X_train, Y_train) and 30% was kept for testing (X_test, Y_test). This helps evaluate how well the model performs on unseen data. A random_state value of 2 was used to ensure the results are reproducible each time the code runs.

## **11. Model Selection and Train**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

## **12. Model Evaluation**

```python
X_train_prediction = model.predict(X_train)
```

```python
X_train_prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_train,X_train_prediction)*100
```

```python
X_test_prediction = model.predict(X_test)
```

```python
accuracy_score(Y_test,X_test_prediction)*100
```

After training the model, its performance was checked using accuracy scores. The model achieved an accuracy of 80% on the training data and 78% on the testing data. This shows that the model has learned well from the data and is also able to make good predictions on new, unseen data.
