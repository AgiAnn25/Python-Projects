```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://loveandflair.com/collections/activewear-1"
headers = {"User-Agent": "Mozilla/5.0"}
```

```python
response = requests.get(url, headers=headers)
```

```python
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
all_classes = []
for tag in soup.find_all():
    if tag.get('class'):
        all_classes.extend(tag.get('class'))
print(set(all_classes))
```

```python
for tag in soup.find_all():
     if tag.get('class'):
        class_names = " ".join(tag.get('class'))
     if 'product' in class_names.lower():
        print(tag.name, tag.get('class'))
```

```python
products = soup.find_all('li', class_='collection-product-card')

print(len(products))
```

```python

print(products[0].prettify())
```

```python
data = []
```

```python
for product in products:
    try:
        name = product.find(class_='card__title').text.strip()
    except:
        name = "N/A"
    try:
        price = product.find(class_='price-item--regular').text.strip()
    except:
        price = "N/A"
    try:
        stock = product.find(class_='sold-out')
        stock = "Sold Out" if stock else "In Stock"
    except:
        stock = "N/A"
    try:
        color = product.find(class_='color-swatch')['title']
    except:
        color = "N/A"
    try:
        product_link = product.find(class_='card-product__link')['href']
        full_link = "https://loveandflair.com" + product_link

        product_response = requests.get(full_link, headers=headers)
        product_soup = BeautifulSoup(product_response.text, 'html.parser')

        size_tags = product_soup.find_all(class_='link-hover-line-outer')

        sizes = ", ".join([
            size.find('span', class_='uppercase').text.strip()
            for size in size_tags
        ])
    except:
        sizes = "N/A"

    data.append({
        "Name": name,
        "Price": price,
        "Stock": stock,
        "Color": color,
        "Size": sizes
    })
```

```python
df = pd.DataFrame(data)
df.head(10)
```
